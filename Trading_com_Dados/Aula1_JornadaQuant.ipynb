### Criar filtro

```python
# Filtra o DataFrame para incluir apenas as linhas onde 'patrliq' (Patrimônio Líquido) é maior que zero e 'divbpatr' (Dívida Bruta sobre Patrimônio) é menor que 0.5
filtrado = df[(df.patrliq > 0) & (df.divbpatr < 0.5)]
```

Essa linha filtra o DataFrame original df para incluir apenas as linhas onde o 'patrliq' (Patrimônio Líquido) é maior que zero e 'divbpatr' (Dívida Bruta sobre Patrimônio) é menor que 0.5.

```python
# Ordena os resultados filtrados com base na coluna 'divbpatr' de forma ascendente (menor para o maior)
divida_bruta_patr = filtrado.sort_values('divbpatr', ascending=True)
```

Aqui, o DataFrame filtrado é ordenado com base na coluna 'divbpatr' (Dívida Bruta sobre Patrimônio) de forma ascendente, ou seja, do menor para o maior valor dessa métrica.

```python
# Exibe as primeiras 50 linhas do DataFrame resultante após as operações
divida_bruta_patr.head(50)
```

```python
import yfinance as yf
from datetime import datetime, timedelta

def is_traded_recently(symbol):
    try:
        # Trazer os dados históricos de cotação para aquela ação nos últimos 1 mês
        stock = yf.Ticker(symbol)
        historical_data = stock.history(period='1mo')

        # Trazer a cotação mais recente e obter a data correspondente
        last_trading_day = historical_data.index[-1].date()

        # Verifica se o dia de negociação mais recente está dentro de uma janela de 5 dias úteis
        today = datetime.now().date()
        window_start = today - timedelta(days=5)

        return window_start <= last_trading_day <= today
    except Exception:
        # Retorna False se houver algum erro ao obter os dados
        return False
```

Essa função is_traded_recently recebe um símbolo de ação como entrada e faz o seguinte:

- Utiliza a biblioteca yfinance para obter dados históricos de cotação para aquela ação nos últimos 1 mês.
- Obtém a data correspondente à cotação mais recente.
- Verifica se o dia de negociação mais recente está dentro de uma janela de 5 dias úteis (window_start <= last_trading_day <= today).
- Retorna True se a ação foi negociada recentemente e False caso contrário.

A exceção (except Exception) trata de erros que possam ocorrer ao tentar obter os dados financeiros. Se houver algum erro, a função retorna False.

### Filtrando apenas papéis negociados

```python
# Adiciona uma nova coluna 'ACAO' ao DataFrame, contendo a concatenação dos índices com a string '.SA'
divida_bruta_patr['ACAO'] = divida_bruta_patr.index + '.SA'
```

Essa linha cria uma nova coluna 'ACAO' no DataFrame divida_bruta_patr. O conteúdo dessa coluna é obtido concatenando os índices existentes (que são códigos de empresas) com a string '.SA'. Isso cria um novo identificador que parece representar códigos de ações listadas na B3.

```python
# Adiciona uma nova coluna 'NEGOCIADA' ao DataFrame, usando a função is_traded_recently na coluna 'ACAO'
divida_bruta_patr['NEGOCIADA'] = divida_bruta_patr['ACAO'].apply(is_traded_recently)
```

Essa linha usa o método apply para aplicar a função is_traded_recently a cada elemento da coluna 'ACAO' do DataFrame divida_bruta_patr. O resultado dessa aplicação é uma nova coluna chamada 'NEGOCIADA', que indica se cada ação foi negociada recentemente (True) ou não (False).

```python
# Seleciona as primeiras 50 linhas do DataFrame divida_bruta_patr só para agilizar o processamento do código
divida_bruta_patr = divida_bruta_patr.head(50)

# Filtra as linhas onde a coluna 'NEGOCIADA' é igual a True, ou seja, onde as ações foram negociadas recentemente
empresas_filtradas = divida_bruta_patr[divida_bruta_patr['NEGOCIADA'] == True]
```

## Encontrando as empresas maiores ROE

```python
# Ordena as empresas filtradas com base na coluna 'roe' (Retorno sobre o Patrimônio Líquido) de forma descendente (do maior para o menor)
empresas_filtradas = empresas_filtradas.sort_values('roe', ascending=False).head(10)
```

```python
# Configura o tamanho da figura do gráfico
plt.figure(figsize=(10, 6))

# Cria um gráfico de barras usando os índices (nomes das empresas) e os valores de ROE
plt.bar(empresas_filtradas.tail(10).index, empresas_filtradas.tail(10).roe)

# Rotaciona os rótulos do eixo x para melhor legibilidade
plt.xticks(rotation=45)

# Adiciona um título ao gráfico
plt.title("Ações selecionadas por Divida Bruta/Pat Liquido e ROE")
```
